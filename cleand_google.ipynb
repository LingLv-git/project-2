```python
import pandas as pd

%load_ext nb_black
```

```python
# File to Load (Remember to Change These)
csv_file = "google_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(csv_file)
df.head()
```

```python
new_df = df[["name", "type", "rating", "reviews", "range"]].copy()
new_df.drop_duplicates(subset=["name"], inplace=True)
restaurant = new_df[new_df["type"].str.contains("aurant", regex=True)]
```

```python
restaurant.to_csv("google_clean_data.csv")
```
